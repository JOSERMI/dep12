```python
df_orders["month"]= df_orders['order_date'].astype(str).str[5:7]
df_orders['month'] = df_orders['month'].astype('int32')
df_orders.head()
```

```python
def month_to_quarter(month): 
    period = pd.Period(month=month, freq='M') 
    quarter = f'Q{period.quarter}' 
    return period.month_name(), quarter
```

```python
def month_to_quarter(month):
    return (month-1)//3 + 1
```

```python
df_orders['order_date'] = pd.to_datetime(df_orders['order_date'])
```

```python
df_orders['Q2'] = df_orders['order_date'].dt.quarter
```

```python
df_orders["Q"]= df_orders['month'].apply(month_to_quarter)
```

```python
df_orders['year'] = df_orders['order_date'].dt.year
```

```python
df_orders.head()
```

```python
df_sort_orders = df_orders.sort_values('order_date', ascending=0)
```

```python
df_sort_orders.head()
```

```python
q = df_sort_orders.iloc[0]['Q']
year = df_sort_orders.iloc[0]['year']
```

```python
df_join  = df_sort_orders.query(f" Q=={q} & year=={year}").merge(df_order_items, left_on='order_id', right_on='order_item_order_id')\
                    .merge(df_products, left_on='order_item_product_id', right_on='product_id')\
                    .merge(df_categories, left_on='product_category_id', right_on='category_id')\
                    .merge(df_departments, left_on='category_department_id', right_on='department_id')
```

```python
df_join = df_join[['department_name', 'product_name', 'order_item_quantity']]
```

```python
df = df_join.groupby(['department_name','product_name']).sum('order_item_quantity').reset_index()
```

```python
df.head(10)
```

```python
df['dense_rank'] = (df.groupby('department_name')['order_item_quantity']
                      .rank(method='dense', ascending=False)
                      .astype(int)
                     )
```

```python
df_ = df.sort_values(['department_name','dense_rank'])
```
